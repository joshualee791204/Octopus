```python
##Load visualization and data frame tools for each column
%matplotlib inline
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import matplotlib
```

```python
raw_df = pd.read_csv('train.csv')
df = raw_df.copy()
```

```python
df.SalePrice.dropna().shape #returns 1460, thus NO NAs
```

```python
df.SalePrice.nunique()# returns 663. Morethan half of the price have duplicates
```

```python
df.SalePrice.describe()
```

```python
#positive skewness means right side actual value is higher than norm. 
print("Skewness: %f" % df['SalePrice'].skew())
#The kurtosis of any univariate normal distribution is 3. This result is leptrutosic or thin(tall)
print("Kurtosis: %f" % df['SalePrice'].kurt())
```

```python
from math import sqrt
print( pd.Series([log(x, 10) for x in df.SalePrice]).skew())
print( pd.Series([log(x, 10) for x in df.SalePrice]).kurt())
```

```python
print( pd.Series([sqrt(x) for x in df.SalePrice]).skew())
print(pd.Series([sqrt(x) for x in df.SalePrice]).kurt())
```
